```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
df = pd.read_csv('data/raw/additional_dwellings.csv')
df.head()
```

```python
plt.hist(x='AppliedDate', bins=10, data=df)
plt.xticks(rotation='vertical');
```

# 23/10/06

I was hoping to find an easy solution for the xticks. However, looks like there is not a super obvious one (or atleast google is hiding it from me). It might be a problem with the `AppliedDate` field. So, the plan is to convert it all to `datetime` objects

```python
df['AppliedDate'] = pd.to_datetime(df['AppliedDate'])
```

```python
plt.hist(x='AppliedDate', bins=12, data=df, color='moccasin')
plt.title('Additional Dwelling Permits Applied for in Calgary')
plt.xlabel('Applied Date')
plt.ylabel('Number of Permits')
plt.savefig('figures/additional_dwellings_hist.png')
```
