```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data
path = 'data/raw/demolition/'
df_2000 = pd.read_csv(path + 'demolition_2000.csv')
df_2005 = pd.read_csv(path + 'demolition_2005.csv')
df_2010 = pd.read_csv(path + 'demolition_2010.csv')
df_2015 = pd.read_csv(path + 'demolition_2015.csv')
```

```python
# preview
display(df_2000)
```

# Total demolition

```python
# Get the number of demolitions in each half-decade
count_2000 = df_2000.shape[0]
count_2005 = df_2005.shape[0]
count_2010 = df_2010.shape[0]
count_2015 = df_2015.shape[0]
```

```python
# Create our series
series = pd.Series([count_2000, count_2005, count_2010, count_2015])
series.rename({0:2000, 1:2005, 2:2010, 3:2015}, inplace=True)
```

```python
series.plot(kind='bar', x='Half-Decade Start', title='Number of Demolition Permits Applied for')
```
